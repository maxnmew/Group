```python
import pandas as pd
import os
from IPython.display import display
# Host path you provided (may not be accessible from the notebook kernel)
host_path = "/Users/maxine/Desktop/Machine_Learning_Degree/vscode/data/umsatzdaten_gekuerzt.csv"
# Workspace fallback path (place file under ./data/ in the repo/workspace)
workspace_path = os.path.join(os.getcwd(), 'data', 'umsatzdaten_gekuerzt.csv')
df = None
sep = ';'  # adjust if your CSV uses a different separator
encoding = 'utf-8'
# Try host path (only works if kernel can access that filesystem)
try:
    host_expanded = os.path.expanduser(host_path)
    if os.path.exists(host_expanded):
        df = pd.read_csv(host_expanded, sep=sep, encoding=encoding)
        print(f"Loaded CSV from host path: {host_expanded}")
    else:
        raise FileNotFoundError(host_expanded)
except Exception as e_host:
    print("Could not read from host path:", e_host)
    # Try workspace path next
    try:
        if os.path.exists(workspace_path):
            df = pd.read_csv(workspace_path, sep=sep, encoding=encoding)
            print(f"Loaded CSV from workspace: {workspace_path}")
        else:
            raise FileNotFoundError(workspace_path)
    except Exception as e_workspace:
        print("Could not read from workspace path:", e_workspace)
        print("File not found. To load the CSV into this notebook:")
        print("  - Copy the file into the workspace 'data/' folder and re-run this cell.")
        print("  - Or upload the file via VS Code Explorer into the project 'data/' folder.")
        print("Example host command (adjust paths):")
        print("  cp /Users/maxine/Desktop/Machine_Learning_Degree/vscode/data/umsatzdaten_gekuerzt.csv /path/to/your/project/data/")
        print("After copying, re-run this cell.")
# If loaded, show a small preview and shape
if df is not None:
    display(df.head())
    print(f"DataFrame shape: {df.shape}")
else:
    print("No DataFrame was loaded.")
```
